```python
def rolling_mad_outlier_detector(x, xerr, threshold=3.0,
                                 window=5, debug=False):
    """ Check whether a point is an outlier or not based on its 
    neighborhood of points and the median absolute deviation of
    that neighborhood.

    """
    
    output_mask = np.zeros(len(x), dtype=bool)
    z_arr = np.zeros(len(x))
    for ind,datapt in enumerate(tqdm(x)):
        m_ind = max(0, ind - int(window/2.))
        p_ind = min(len(x)-1, ind + int(window/2.))

        neighborhood = x[m_ind:p_ind]
        neighborhood_e = xerr[m_ind:p_ind]
        
        weighted_median = wq.median(neighborhood, 1. / neighborhood_e**2.)
        
        MAD = stats.median_abs_deviation(neighborhood)
        
        modified_sd = 1.4826 * MAD # normalization constant to bring MAD to SD
        
        modified_z = abs(x[ind] - weighted_median) / (modified_sd**2. + xerr[ind]**2.)**.5
        z_arr[ind] = modified_z
        output_mask |= modified_z > threshold
#         print(modified_z)
    if debug:
        return output_mask, z_arr
    return output_mask
```

```python
def two_neighbor_outlier_detector(x, xerr, threshold=3.0,
                                 debug=False):
    """ Compute normalized euclidean distance and return True/False based
    on each datapoints distance to it's front and back neighbors in 1D
    
    """
    
    output_mask = np.zeros(len(x), dtype=bool)
    sig_vals = np.ndarray((len(x),2))
    
    for ind,datapt in enumerate(tqdm(x)):
        m_ind = max(0, ind - 1)
        p_ind = min(len(x)-1, ind + 1)

        m_pt, m_unc = x[m_ind], xerr[m_ind]
        p_pt, p_unc = x[p_ind], xerr[p_ind]
        
        m_diff = abs(m_pt - x[ind])
        p_diff = abs(p_pt - x[ind])
        
        m_z = m_diff / (m_unc**2. + xerr[ind]**2.)**.5
        p_z = p_diff / (p_unc**2. + xerr[ind]**2.)**.5
        
        m_mask = m_z > threshold
        p_mask = p_z > threshold
        
        if debug:
            sig_vals[ind][0] = m_z
            sig_vals[ind][1] = p_z
        output_mask |= m_mask*p_mask == 1
    if debug:
        return output_mask, sig_vals
    return output_mask
```
